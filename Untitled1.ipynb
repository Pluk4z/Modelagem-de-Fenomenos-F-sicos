```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x + np.sin(3*x)

def df_fd(x, h):
    return (f(x+h) - f(x))/h

def df_mid(x, h):
    return (f(x+h/2) - f(x-h/2))/h

x = np.arange(0, 2*np.pi, 0.01)
y_exact = 2 + 3*np.cos(3*x) # derivada exata
y_fd = df_fd(x, 1e-6) # derivada por diferenças finitas
y_mid = df_mid(x, 1e-6) # derivada por ponto médio

plt.plot(x, y_exact, label='Derivada exata')
plt.plot(x, y_fd, label='Diferenças finitas')
plt.plot(x, y_mid, label='Ponto médio')
plt.legend()
plt.show()

x0 = np.pi/4
h = 1e-6

y_exact_x0 = -3*np.sin(3*x0) + 2
y_fd_x0 = df_fd(x0, h)
y_mid_x0 = df_mid(x0, h)

error_fd = abs(y_exact_x0 - y_fd_x0)
error_mid = abs(y_exact_x0 - y_mid_x0)

print(f'Erro na Fórmula das Diferenças Finitas: {error_fd:.20f}')
print(f'Erro na Fórmula do Ponto Médio: {error_mid:.20f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x + np.sin(3*x)

def df_fd(x, h):
    return (f(x+h) - f(x))/h

def df_mid(x, h):
    return (f(x+h/2) - f(x-h/2))/h

def dy_exact(x):
    return -3*np.sin(3*x) + 2

def d2y_exact(x):
    return -27*np.cos(3*x)

def newton_raphson(f, df, x0, tol=1e-12, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        if abs(fx) < tol:
            return x
        x -= fx/dfx
    return x

x = np.arange(0, 2*np.pi, 0.01)
y_exact = 2 + 3*np.cos(3*x) # derivada exata
y_fd = df_fd(x, 1e-6) # derivada por diferenças finitas
y_mid = df_mid(x, 1e-6) # derivada por ponto médio

plt.plot(x, y_exact, label='Derivada exata')
plt.plot(x, y_fd, label='Diferenças finitas')
plt.plot(x, y_mid, label='Ponto médio')
plt.legend()
plt.show()

x0 = np.pi/3
h = 1e-6

y_exact_x0 = dy_exact(x0)
y_fd_x0 = df_fd(x0, h)
y_mid_x0 = df_mid(x0, h)

error_exact_fd = abs(y_exact_x0 - y_fd_x0)
error_exact_mid = abs(y_exact_x0 - y_mid_x0)
error_fd_mid = abs(y_fd_x0 - y_mid_x0)

print(f'Erro na Fórmula das Diferenças Finitas: {error_fd:.20f}')
print(f'Erro na Fórmula do Ponto Médio: {error_mid:.20f}')
print(f'Erro entre derivada exata e Fórmula das Diferenças Finitas: {error_exact_fd:.20f}')
print(f'Erro entre derivada exata e Fórmula do Ponto Médio: {error_exact_mid:.20f}')
print(f'Erro entre Fórmula das Diferenças Finitas e Fórmula do Ponto Médio: {error_fd_mid:.20f}')
```
