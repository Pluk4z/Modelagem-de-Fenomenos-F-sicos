# Questão 1: 
O valor de um computador deprecia exponencialmente com o tempo. Seu valor inicial é de R$ 4.000, e a taxa de depreciação anual é de 20%. Qual será o valor do computador após 5 anos? Use a fórmula:

$$
V(t) = V_0 \cdot (1 - r)^t
$$

onde $V_0$ é o valor inicial, $r$ é a taxa de depreciação e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o o valor do bem depreciado, ano a ano, nos próximos 15 anos.

```python
import matplotlib.pyplot as plt

v = 4000
r = 0.2
anos = 15

anos_lista = range(anos + 1)
valor_lista = [v*(1-r)**t for t in anos_lista]

print(f'Valor do computador após 5 anos = {valor_lista[5]}')

plt.plot(anos_lista, valor_lista)
plt.title('Grafico')
plt.xlabel('anos')
plt.ylabel('Valor')

plt.show()
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $3^{x-1} - 2^x = 7$

2. $\log_5(2x + 1) + \log_5(x - 2) = 2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Função 1

x = np.linspace(-5, 5, 1000)
y = 3**(x-1) - 2**x - 7

plt.plot(x, y)
plt.title(r"$3^{x-1} - 2^x = 7$, crescente")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


# Função 2

x = np.linspace(2, 5, 1000)
y = np.log(2*x+1)/np.log(5) + np.log(x-2)/np.log(5) - 2

plt.plot(x, y)
plt.title(r"$\log_5(2x+1) + \log_5(x-2) = 2$, crescente")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 0} \frac{\sin(2x)}{x}$

2. $\lim_{x \to \infty} \frac{e^x}{x^3}$

Começando com o primeiro limite:

$\lim_{x \to 0} \frac{\sin(2x)}{x}$

Podemos utilizar a regra de L'Hôpital, que afirma que o limite de uma função f(x) sobre g(x) é igual ao limite da derivada de f(x) sobre a derivada de g(x), desde que este limite exista e seja finito ou infinito. Nesse caso, podemos aplicar essa regra uma vez, já que temos a forma indeterminada 0/0:

$\lim_{x \to 0} \frac{\sin(2x)}{x}$

(aplicando a regra de L'Hôpital)

$\lim_{x \to 0} \frac{2 cos(2x)}{1} = 2$


Portanto, o limite é igual a 2.

```python
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = sp.sin(2*x)/x

sp.plot(f, (x, -5, 5), ylim=(-2, 2))
plt.show()
```

$\lim_{x \to \infty} \frac{e^x}{x^3}$

Nesse caso, podemos utilizar a regra de L'Hôpital novamente, mas antes precisamos simplificar a expressão utilizando a propriedade da exponenciação:

$\lim_{x \to \infty} \frac{e^x}{x^3} = lim_{x \to \infty} {e^{(x - 3\ln (x))}}$


Agora podemos aplicar a regra de L'Hôpital, já que temos a forma indeterminada $\frac{\infty}{\infty}$:

$\lim_{x \to \infty} \frac{e^{(x - 3\ln (x))}}{3x^{-2}}$


$\lim_{x \to \infty} \frac{e^x}{x^3} = lim_{x \to \infty} \frac{e^{(x - 3\ln (x))}}{6x^{-2}}$ (aplicando a regra de L'Hôpital)

$\lim_{x \to \infty}\frac{e^{(x - 3\ln (x))}}{18} $

Como x cresce sem limite quando $x \to \infty$, a expressão $x - 3\ln (x)$ cresce mais rápido do que $x$, então ${e^{(x - 3\ln (x))}}$ cresce mais rápido do que qualquer potência de $x$, o que implica que o limite é infinito.

```python

import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = sp.exp(x)/x**3

sp.plot(f, (x, -5, 5), ylim=(-10, 10))
plt.show()
```
